```python
def load_and_clean_irs(filepath, year):
    # Load the file skip the title rows
    df = pd.read_excel(filepath, header=None, skiprows=10)

    # Keep only the 12 bracket rows
    df = df.iloc[0:12]

    # Extract the columns
    df_clean = df[[0, 1, 2, 3]].copy()

    # Rename column
    df_clean.columns = ['income_bracket', 'num_returns', 'agi', 'total_income_tax']

    # Replace * and ** with NaN
    df_clean = df_clean.replace(['*', '**'], pd.NA)

    # Compute effective tax rate
    df_clean['effective_tax_rate'] = df_clean['total_income_tax'] / df_clean['agi']

    # Add year
    df_clean['year'] = year

    return df_clean
```

```python
# test
df_2022 = load_and_clean_irs('../data/raw/irs_soi_table32_2022.xls', 2022)
df_2022
```

```python
# Load all 6 function
df_32_2021 = load_and_clean_irs('../data/raw/irs_soi_table32_2021.xls', 2021)
df_32_2022 = load_and_clean_irs('../data/raw/irs_soi_table32_2022.xls', 2022)
df_32_2023 = load_and_clean_irs('../data/raw/irs_soi_table32_2023.xls', 2023)

# Check they all loaded correctly
print("2021 shape:", df_32_2021.shape)
print("2022 shape:", df_32_2022.shape)
print("2023 shape:", df_32_2023.shape)
```

```python
import pandas as pd

#  all 3 years together
df_table32_all = pd.concat([df_32_2021, df_32_2022, df_32_2023], ignore_index=True)

# Verify (12 x 3 years)
print("Combined shape:", df_table32_all.shape)
df_table32_all
```

```python
# all 3 years of Table 3.3 (filing status)
df_33_2021 = load_and_clean_irs('../data/raw/irs_soi_table33_2021.xls', 2021)
df_33_2022 = load_and_clean_irs('../data/raw/irs_soi_table33_2022.xls', 2022)
df_33_2023 = load_and_clean_irs('../data/raw/irs_soi_table33_2023.xls', 2023)

# Check shapes
print("2021 shape:", df_33_2021.shape)
print("2022 shape:", df_33_2022.shape)
print("2023 shape:", df_33_2023.shape)
```

```python
#  3 years of Table 3.3
df_table33_all = pd.concat([df_33_2021, df_33_2022, df_33_2023], ignore_index=True)

# Verify 36 rows
print("Combined shape:", df_table33_all.shape)
df_table33_all
```

```python
# Combine Table 3.2 and Table 3.3  master dataset
df_master = pd.concat([df_table32_all, df_table33_all], ignore_index=True)

# 72 rows (36 + 36)
print("Master dataset shape:", df_master.shape)

# Save
df_master.to_csv('../data/processed/irs_master_clean.csv', index=False)

print(" Master CSV saved to data/processed/irs_master_clean.csv")
```
