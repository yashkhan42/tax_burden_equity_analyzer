```python
import pandas as pd

# Load one IRS file to explore
df = pd.read_excel('../data/raw/irs_soi_table32_2022.xls',
                   header=None,
                   skiprows=10)

# See what we're working with
print(df.shape)
df.head(20)
```

```python
# See all 16 rows to identify what needs to be dropped
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 20)
df
```

```python
# Drop the total and footnote rows, keep only the 12 bracket rows
df_clean = df.iloc[0:12]

# Verify we have exactly 12 rows
print(df_clean.shape)
df_clean
```

```python
# See all column names/numbers
for i, col in enumerate(df_clean.columns):
    print(f"Column {i}: {col}")
```

```python
# Look at specific columns with full numbers
pd.set_option('display.float_format', '{:.2f}'.format)
df_clean.iloc[:, 0:5]
```

```python
# Look at just the first 5 columns to identify them
df_clean.iloc[:, 0:5]
```

```python
# Extract just the columns we need and rename them
df_working = df_clean[[0, 1, 2, 3]].copy()

# Give them proper names
df_working.columns = ['income_bracket', 'num_returns', 'agi', 'total_income_tax']

# Compute the effective tax rate
df_working['effective_tax_rate'] = df_working['total_income_tax'] / df_working['agi']

# Add the year
df_working['year'] = 2022

# See the result
df_working
```
